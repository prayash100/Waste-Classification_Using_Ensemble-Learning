```python
import shutil
import os

# Define the paths of the folders you want to exclude
source_dir = '/content'  # Replace with your source directory if different
exclude_folders = ['/content/waste-seg-400-1', '/content/drive']  # Folders to exclude

# Temporary directory for zipping without the excluded folders
temp_dir = '/content/temp'

# Create the temp directory if it doesn't exist
os.makedirs(temp_dir, exist_ok=True)

# Loop through the source directory and copy everything except the excluded folders and `.gdoc` files
for foldername in os.listdir(source_dir):
    folder_path = os.path.join(source_dir, foldername)
    if folder_path not in exclude_folders:
        if os.path.isdir(folder_path):
            shutil.copytree(folder_path, os.path.join(temp_dir, foldername), ignore=shutil.ignore_patterns('*.gdoc'))
        elif not folder_path.endswith('.gdoc'):  # Skip `.gdoc` files
            shutil.copy(folder_path, temp_dir)

# Create a zip of the new directory
shutil.make_archive('/content/downloaded_files', 'zip', temp_dir)

# Download the zip file
from google.colab import files
files.download('/content/downloaded_files.zip')
```
